Importing the libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data collection and analysis 


```python
#loading the data from csv file to PANDAS dataframe 
parki_data=pd.read_csv('/content/parkinsons.csv')
```

Data Pre-Processing 

Separating the features & Target 


```python
X= parki_data.drop(columns=['name','status'],axis=1)
Y= parki_data['status']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data to training data &Test data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train= scaler.transform(X_train)

X_test= scaler.transform(X_test)
```

```python
print(X_train)
```

Model Training  

Support Vector Machine Model

```python
model = svm.SVC(kernel='linear')
```

```python
#training the SVM model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy score

```python
#accuracy score on training data 
X_train_prediction= model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
```

```python
print('Accuracy score of training data :',training_data_accuracy)
```

```python
#accuracy score on training data 
X_test_prediction= model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test,X_test_prediction)
```

```python
print('Accuracy score of test data :',test_data_accuracy)
```

Building a predictive system

```python
input_data=(162.56800,198.34600,77.63000,0.00502,0.00003,0.00280,0.00253,0.00841,0.01791,0.16800,0.00793,0.01057,0.01799,0.02380,0.01170,25.67800,0.427785,0.723797,-6.635729,0.209866,1.957961,0.135242)

#checking input data to a numpy array 
input_data_as_numpy_array = np.asarray(input_data)

#reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

#standardize the data 
std_dta= scaler.transform(input_data_reshaped)

prediction = model.predict(std_dta)
print(prediction)

if (prediction[0] == 0):
  print("The person does not have Parkinsons Disease")
else:
  print("The person has Parkinsons")
```
